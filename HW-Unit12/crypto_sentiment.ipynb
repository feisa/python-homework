# News Headlines Sentiment

Use the news api to pull the latest news articles for bitcoin and ethereum and create a DataFrame of sentiment scores for each coin. 

Use descriptive statistics to answer the following questions:
1. Which coin had the highest mean positive score?
2. Which coin had the highest negative score?
3. Which coin had the highest positive score?

```python
# Initial imports
import os
import pandas as pd
from dotenv import load_dotenv
from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
from newsapi import NewsApiClient

%matplotlib inline
```

```python
# Read your api key environment variable
from dotenv import load_dotenv
load_dotenv()
api_key = os.getenv("news_api")
```

```python
# Create a newsapi client
newsapi = NewsApiClient(api_key=os.environ["news_api"])
```

```python
# Fetch the Bitcoin news articles
btc_headlines = newsapi.get_everything(
    q='bitcoin',
    language='en',
    sort_by='relevancy'
)
```

```python
# Fetch the Ethereum news articles
eth_headlines = newsapi.get_everything(
    q='ethereum',
    language='en',
    sort_by='relevancy'
)
```

```python
# Create the Bitcoin sentiment scores DataFrame
sentiments = []

for articles in btc_headlines['articles']:
    try:
        text = articles['content']
        results = analyzer.polarity_scores(text)
        compound = results['compound']
        pos = results['pos']
        neu = results['neu']
        neg = results['neg'] 

        #put data into sentiments
        sentiments.append({
            'text':text,
            'Compound':compound,
            'Positive':pos,
            'Negative':neg,
            'Neutral':neu})

    except AttributeError:
        pass

btc = pd.DataFrame(sentiments)
btc.head()
```

```python
# Create the ethereum sentiment scores DataFrame
sentiments = []

for articles in eth_headlines['articles']:
    try:
        text = articles['content']
        results = analyzer.polarity_scores(text)
        compound = results['compound']
        pos = results['pos']
        neu = results['neu']
        neg = results['neg'] 

        #put data into sentiments
        sentiments.append({
            'text':text,
            'Compound':compound,
            'Positive':pos,
            'Negative':neg,
            'Neutral':neu})

    except AttributeError:
        pass

eth = pd.DataFrame(sentiments)
eth.head()
```

# Tokenizer

In this section, you will use NLTK and Python to tokenize the text for each coin. Be sure to:
1. Lowercase each word
2. Remove Punctuation
3. Remove Stopwords

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from string import punctuation
import re
```

```python
# Expand the default stopwords list if necessary
wnl = WordNetLemmatizer()

stop = stopwords.words('english')

stop.append('``')
stop.append('\'')
stop.append('...')
stop = set(stop)
```

```python
# Complete the tokenizer function
def tokenizer(text):
    words = word_tokenize(text)
    words = list(filter(lambda w: w.lower(), words))
    words = list(filter(lambda t: t not in punctuation, words))
    words = list(filter(lambda t: t.lower() not in stop, words))
    
    tokens = [wnl.lemmatize(word) for word in words]
    return tokens
```

```python
btc['tokens'] = btc.text.apply(tokenizer)
btc.head()
```

```python
eth['tokens'] = eth.text.apply(tokenizer)
eth.head()
```

# NGrams and Frequency Analysis

In this section you will look at the ngrams and word frequency for each coin. 

1. Use NLTK to produce the n-grams for N = 2. 
2. List the top 10 words for each coin. 

```python
from collections import Counter
from nltk import ngrams
```

```python
# Generate the Bitcoin N-grams where N=2
N = 2
btc_grams = ngrams(tokenizer(btc.text.str.cat()), N)
Counter(btc_grams).most_common(20)
```

```python
# Generate the Ethereum N-grams where N=2
eth_grams = ngrams(tokenizer(eth.text.str.cat()), N)
Counter(eth_grams).most_common(20)
```

```python
# Use the token_count function to generate the top 10 words from each coin
def token_count(token, N=3):
    return Counter(token).most_common(N)
```

```python
# Get the top 10 words for Bitcoin
btc_tokens = tokenizer(btc.text.str.cat())
token_count(btc_tokens, 10)
```

```python
# Get the top 10 words for Ethereum
eth_tokens = tokenizer(eth.text.str.cat())
token_count(eth_tokens, 10)
```

# Named Entity Recognition

In this section, you will build a named entity recognition model for both coins and visualize the tags using SpaCy.

```python
import spacy
from spacy import displacy
```

```python
# Optional - download a language model for SpaCy
!python -m spacy download en_core_web_sm
```

```python
# Load the spaCy model
nlp = spacy.load('en_core_web_sm')
```

## Bitcoin NER

```python
# Concatenate all of the bitcoin text together
all_btc_text = btc.text.str.cat()
all_btc_text
```

```python
# Run the NER processor on all of the text
btc_doc = nlp(all_btc_text)

# Add a title to the document
btc_doc.user_data['title'] = 'BTC Recognition'
```

```python
# Render the visualization
displacy.render(btc_doc, style='ent', jupyter=True)
```

```python
# List all Entities
for ent in btc_doc.ents:
    print(ent.text, ent.label_)
```
